```python
df = df[['Sub-product', 'Consumer complaint narrative']]
df = df.dropna(subset=['Sub-product', 'Consumer complaint narrative'])
df = df.reset_index(drop=True)
print(df.isna().sum())
print(df.shape)
print(df.head(5))
```

```python
top_products = df['Sub-product'].value_counts().nlargest(10).index
df = df[df['Sub-product'].isin(top_products)]
```

```python
df.shape
```

sTRATIFIED SAMPLING TO KEEP THE SAME PROPORTION OF VALUES AS IT WAS

```python
df_reduced = (
    df.groupby('Sub-product', group_keys=False)
      .apply(lambda x: x.sample(frac=600000/len(df), random_state=42))
)

```

```python
df_reduced.shape
```

```python
df_reduced['Sub-product'].value_counts()
```

```python
df_reduced['Sub-product'].nunique()
```

```python
df_reduced['Sub-product'].unique()
```

```python
label_map = {
    "Checking account": "Bank Account",
    "Domestic (US) money transfer": "Bank Account",

    "Credit card debt": "Credit Card",
    "General-purpose credit card or charge card": "Credit Card",

    "Conventional home mortgage": "Mortgage",

    "Federal student loan servicing": "Loan",
    "Loan": "Loan",
    "Other debt": "Loan",

    "Credit reporting": "Credit Reporting",

    "I do not know": "Don't know"
}
```

```python
df_reduced['Department'] = df_reduced['Sub-product'].map(label_map)
```

```python
df_reduced['Department'].value_counts()
```
